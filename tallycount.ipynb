```python
import pandas as pd
```

```python
df = pd.read_csv('SAMC_Scholarly_Activity_Form_Submissions_2025-02-18.csv')
df.info()
```

```python
df_yes = df[df['Will you submit this project for the upcoming SAMC\'s Annual Research Poster Day?'] == 'YES']
df_yes.info()
```

# Total Poster Prelim

```python
count_yes = df['Will you submit this project for the upcoming SAMC\'s Annual Research Poster Day?'].value_counts().get('YES', 0)
print(count_yes)
```

## Emergency Med

```python
count_emergency_yes = df[(df.iloc[:, 8] == 'Emergency Medicine') & (df.iloc[:, 84] == 'YES')].shape[0]
print(count_emergency_yes)
```

## Family Med

```python
count_family_medicine_yes = df_yes[df_yes['Which Department do you belong to? (Please choose ONE)'] == 'Family Medicine'].shape[0]
print(count_family_medicine_yes)
```

### Columns for IM (PI name, Other members, Title, Research Type, and Research Day)

```python
# Assuming df is your DataFrame
filtered_df = df[(df.iloc[:, 8] == 'Internal Medicine') & (df.iloc[:, 84] == 'YES')]

# Select the specified columns (4, 5, 3, 8, 14, 84)
selected_columns = filtered_df.iloc[:, [4, 5, 3, 8, 14, 84]]

# Sort the DataFrame by the column at index 4 in ascending order
sorted_df = selected_columns.sort_values(by=selected_columns.columns[4], ascending=True)

# Re-index the DataFrame starting from 1
sorted_df.index = range(1, len(sorted_df) + 1)

# Display the new DataFrame
sorted_df
```

```python
sorted_df.to_csv('sorted_df_IM_2025-02-18.csv', index=True)
```
